```python
import os
import medmnist
from medmnist import BloodMNIST

"""

Run this cell once to download the BloodMNIST dataset

- change data_folder path if needed

"""

data_folder = "/blue/bme6938/gael.garcia/bme6938-project-2/data"
data_file = os.path.join(data_folder, "bloodmnist_224.npz")

# only download if file doesn't exist yet
should_download = not os.path.exists(data_file)

train_dataset = BloodMNIST(split='train', download=should_download, size=64, root=data_folder)
val_dataset   = BloodMNIST(split='val',   download=should_download, size=64, root=data_folder)
test_dataset  = BloodMNIST(split='test',  download=should_download, size=64, root=data_folder)

print(f"Train: {len(train_dataset)} | Val: {len(val_dataset)} | Test: {len(test_dataset)}")
```
